## Loading data from city
### Default settings
Create `CityMapper` class instance

```python
cm = CityMapper.CityMapper()
```

`load_data_from_city()` with default parameters loads city footprint, roads, cycleways, water, and railways.

```python
cm.load_data_from_city(city_name = 'Pittsburgh, Pennsylvania')
```

`plot_map()` with default parameters generates a map of downloaded data and returns matplotlib `fig`.

```python
fig = cm.plot_map()
```

`savefig()` saves pdf and png of `fig`

```python
cm.savefig(fig)
```

### Using optional parameters

Add road-to-cycleway ratio with `road_cycleway_ratio_subtitle` set to True. Roads and cycleways must be loaded.

```python
fig = cm.plot_map(
    road_cycleway_ratio_subtitle = True
)
```

Use `city_elements` dictionary to choose what elements to load. Options are `roads`, `cyclways`, `water`, `green`, and `buildings`.

```python
cm.load_data_from_city(
    city_name = 'Pittsburgh, Pennsylvania',
    city_elements = {
        'cycleways': False, 'green': True
    }
)
```

Change the colors of the elements in the map with `colors` dict. Use matplotlib color names. Or change to `None` to omit from map.  
Change the edge colors of the elements in the map with the `edge_colors` dict.  
Change the width of the edges with the `edge_widths` dict.

```python
fig = cm.plot_map(
    colors = {
        'green': 'olivedrab',
        'water': 'steelblue',
        'city_area': None
    },
    edge_colors = {
        'green': 'darkolivegreen',
        'water': 'lightsteelblue',
        'city_area': 'darkgrey'
    },
    edge_width = {
        'railways': .25
    }
)
```

When `city_limits` is `False`, all elements within the bounding rectangle of the city limits are loaded.

```python
cm.load_data_from_city(
    city_name = 'Pittsburgh, Pennsylvania',
    city_limits = False
)
fig = cm.plot_map()
```

# Load data from neighborhood

```python
cm = CityMapper.CityMapper()
cm.load_data_from_city(
    'Back Bay, Boston, Massachusetts',
    city_limits=False,
    city_elements={'cycleways':False,'buildings':True,'green': True})

fig = cm.plot_map(edge_width={'roads':1, 'railways': .5})
```

# Load data from address
Set `address` as focal point.
Use `distance` (in meters) to determine how far out map should extend.

```python
cm = CityMapper.CityMapper()
cm.load_data_from_address(
    '1700 E 56th St, Chicago, IL', 
    city_elements={'water':True, 'cycleways':False, 'buildings':True, 'green': True},
    distance = 2000)

fig = cm.plot_map(edge_width={'roads': 1, 'water': .5, 'railways':.25})
```
